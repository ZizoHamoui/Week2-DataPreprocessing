```python
stop_words = stopwords.words('english')
stemmer = SnowballStemmer('english')
```

```python
def clean(text):
    text = text.lower()
    text = re.sub('[^\w\s]', '', text)
    text = [word for word in text.split() if word not in stop_words and word != 'go']
    text = [stemmer.stem(w) for w in text]
    return text
```

```python
dataset_3['text'] = dataset_3['content'].apply(lambda x: clean(x))
dataset_3['text'] = dataset_3['text'].apply(lambda x: ' '.join(x))
dataset_3.head()
```

# Audio Data

Files This portion of the RAVDESS contains 1440 files: 60 trials per actor x 24 actors = 1440. The RAVDESS contains 24 professional actors (12 female, 12 male), vocalizing two lexically-matched statements in a neutral North American accent. Speech emotions includes calm, happy, sad, angry, fearful, surprise, and disgust expressions. Each expression is produced at two levels of emotional intensity (normal, strong), with an additional neutral expression.

### File naming convention

Each of the 1440 files has a unique filename. The filename consists of a 7-part numerical identifier (e.g., 03-01-06-01-02-01-12.wav). These identifiers define the stimulus characteristics:

### Filename identifiers

Modality (01 = full-AV, 02 = video-only, 03 = audio-only).

Vocal channel (01 = speech, 02 = song).

Emotion (01 = neutral, 02 = calm, 03 = happy, 04 = sad, 05 = angry, 06 = fearful, 07 = disgust, 08 = surprised).

Emotional intensity (01 = normal, 02 = strong). NOTE: There is no strong intensity for the 'neutral' emotion.

Statement (01 = "Kids are talking by the door", 02 = "Dogs are sitting by the door").

Repetition (01 = 1st repetition, 02 = 2nd repetition).

Actor (01 to 24. Odd numbered actors are male, even numbered actors are female).

Filename example: 03-01-06-01-02-01-12.wav

1. Audio-only (03)
2. Speech (01)
3. Fearful (06)
4. Normal intensity (01)
5. Statement "dogs" (02)
6. 1st Repetition (01)
7. 12th Actor (12) Female, as the actor ID number is even.

```python
def check_and_plot(audio_path):
    global male_files, female_files, male_count, female_count  # declare at the start
    
    audio_files = glob.glob(os.path.join(audio_path, "*.wav"))
    
    # reset globals each run
    male_files, female_files = [], []
    male_count, female_count = 0, 0

    for f in audio_files:
        base = os.path.splitext(os.path.basename(f))[0]
        last_token = base.split("-")[-1]
        try:
            last_num = int(last_token)
            if last_num % 2 == 1:
                male_files.append(f)
                male_count += 1
            else:
                female_files.append(f)
                female_count += 1
        except ValueError:
            print("⚠️ Could not parse:", f)

    # Plot totals
    plt.bar(["Male", "Female"], [male_count, female_count], color=["blue", "pink"])
    plt.ylabel("Total Files")
    plt.title("Male vs Female Audio Files Count")
    plt.show()

    print("Male:", male_count, "Female:", female_count)
```

```python
check_and_plot("/Users/hamoui/Downloads/Audio Data/Actor_*")
```

## What causes the difference?

```python
def gender_token_analysis(audio_path):
    audio_files = glob.glob(os.path.join(audio_path, "*.wav"))

    records = []

    for f in audio_files:
        base = os.path.splitext(os.path.basename(f))[0]
        tokens = base.split("-")

        if len(tokens) != 7:
            continue 

        try:
            actor_num = int(tokens[-1])
            gender = "Male" if actor_num % 2 == 1 else "Female"

            records.append({
                "gender": gender,
                "modality": tokens[0],
                "vocal_channel": tokens[1],
                "emotion": tokens[2],
                "intensity": tokens[3],
                "statement": tokens[4],
                "repetition": tokens[5],
                "actor": tokens[6],
            })
        except Exception as e:
            print("⚠️ Could not parse:", f, e)

    df = pd.DataFrame(records)

    # Count per gender & token
    summary = {}
    for col in ["modality", "vocal_channel", "emotion", "intensity", "statement", "repetition"]:
        counts = df.groupby(["gender", col]).size().unstack(fill_value=0)
        summary[col] = counts

    return summary
```

```python
summary = gender_token_analysis("/Users/hamoui/Downloads/Audio Data/Actor_*")
```

```python
for token, table in summary.items():
    print(f"\n===== {token.upper()} =====")
    print(table)
    print("Difference (Male - Female):")
    print(table.loc["Male"] - table.loc["Female"])
```
